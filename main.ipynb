## Load TTS model (ESPnet)

- 마키세 크리스 모델 https://huggingface.co/mio/amadeus
- 다른 ESPnet 모델 https://huggingface.co/models?library=espnet

````python
import soundfile
from espnet2.bin.tts_inference import Text2Speech

text2speech = Text2Speech.from_pretrained('mio/amadeus')

print('TTS model loaded!')
````

## Inference TTS

### Example

```
- はじめまして。 처음 뵙겠습니다
- 過去を変えるなんて、なんか卑怯な感じがするから。私の人生はたかだか18年だけど、改変なんかしたくない。失敗したことも含めて、自分の人生なんだから。 과거를 바꿀 수 있다니, 왠지 비겁하다는 생각이 들거든. 나 인생이 기껏해야 18년이지만, 개변따위 하고싶지 않아. 실패한 것도 포함해서, 내 인생이니까.
- 科学なんて99%はつまらないものの積み重ね。 과학이란 99퍼센트는 시시한 것이 쌓인 것이야.
```

````python
from IPython.display import HTML, Audio
from IPython.core.display import display

text = '過去を変えるなんて、なんか卑怯な感じがするから。私の人生はたかだか18年だけど、改変なんかしたくない。失敗したことも含めて、自分の人生なんだから。' #@param {type:"string"}
input_audio_path = '/content/out.wav' #@param {type:"string"}

speech = text2speech(text)['wav']
soundfile.write('/content/out.wav', speech.numpy(), text2speech.fs, 'PCM_16')

display(Audio(input_audio_path))
````

## Inference lip sync model (Wav2Lip)

- pads: Padding (top, bottom, left, right). Please adjust to include chin at least (default. [0, 10, 0, 0])
- rescaleFactor: Reduce the resolution by this factor. Sometimes, best results are obtained at 480p or 720p (default. 1)
- nosmooth: Prevent smoothing face detections over a short temporal window (default. True)
- use_hd_model: Output quality (default. False)

````python
output_file_path = '/content/result_video.mp4' #@param {type:"string"}
pad_top = 0 #@param {type:"integer"}
pad_bottom = 20 #@param {type:"integer"}
pad_left = 0 #@param {type:"integer"}
pad_right = 0 #@param {type:"integer"}
rescaleFactor = 1 #@param {type:"integer"}
nosmooth = True #@param {type:"boolean"}
use_hd_model = True #@param {type:"boolean"}

checkpoint_path = 'checkpoints/wav2lip.pth' if not use_hd_model else 'checkpoints/wav2lip_gan.pth'

if nosmooth == False:
    !python inference.py --outfile $output_file_path --checkpoint_path $checkpoint_path --face $input_video_path --audio $input_audio_path --pads $pad_top $pad_bottom $pad_left $pad_right --resize_factor $rescaleFactor
else:
    !python inference.py --outfile $output_file_path --checkpoint_path $checkpoint_path --face $input_video_path --audio $input_audio_path --pads $pad_top $pad_bottom $pad_left $pad_right --resize_factor $rescaleFactor --nosmooth
````
